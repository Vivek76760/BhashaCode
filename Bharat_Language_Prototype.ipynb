```python
import re

# १. शब्दकोश (The Multi-Lingual Keyword Dictionary)
# Maps regional words to universal mathematical operations
KEYWORDS = {
    # Hindi / Devanagari
    "मान": "LET",
    "यदि": "IF",
    "दिखाओ": "PRINT",

    # Tamil
    "மாறி": "LET",
    "எனில்": "IF",
    "அச்சிடு": "PRINT",

    # Telugu
    "రఖో": "LET",
    "అయితే": "IF",
    "చూపించు": "PRINT"
}

def running_bharat_compiler(script_text):
    memory = {}
    lines = script_text.strip().split('\n')

    for line in lines:
        line = line.strip()
        if not line or line.startswith("#"):
            continue

        # Split the line into words to translate keywords to internal tokens
        words = line.split()
        translated_words = [KEYWORDS.get(w, w) for w in words]
        translated_line = " ".join(translated_words)

        # Action A: Variable Assignment (LET variable = number)
        if translated_line.startswith("LET"):
            match = re.match(r"LET\s+(\w+)\s*=\s*(\d+)", translated_line)
            if match:
                var_name, var_val = match.groups()
                memory[var_name] = int(var_val)

        # Action B: Conditionals & Output (IF var1 < var2 PRINT "message")
        elif translated_line.startswith("IF"):
            match = re.match(r"IF\s+(\w+)\s*([<|>|==]+)\s*(\w+)\s+PRINT\s*\"(.*)\"", translated_line)
            if match:
                var1, op, var2, message = match.groups()
                val1 = memory.get(var1, 0)
                val2 = memory.get(var2, 0)

                # Check condition using standard math rules
                if eval(f"{val1} {op} {val2}"):
                    print(message)

print("Engine Activated and Ready!")
```

```python
# --- टेस्ट १: हिंदी कोड ---
hindi_code = """
मान क = १०
मान ख = २०
यदि क < ख दिखाओ "ख संख्या क से बड़ी है!"
"""

print("--- Running Hindi Script ---")
running_bharat_compiler(hindi_code)


# --- टेस्ट २: தமிழ் கோடு (Tamil Code) ---
# Notice it uses the exact same variables 'क' and 'ख' stored in memory!
tamil_code = """
மாறி क = १०
மாறி ख = २०
எனில் क < ख அச்சிடு "க-வை விட ఖ பெரியது!"
"""

print("\n--- Running Tamil Script ---")
running_bharat_compiler(tamil_code)
```

```python
import re

# Upgraded Word-to-Token Map
KEYWORDS = {
    "मान": "LET", "மாறி": "LET", "రఖో": "LET",
    "यदि": "IF", "எனில்": "IF", "అయితే": "IF",
    "दिखाओ": "PRINT", "அச்சிடு": "PRINT", "చూపించు": "PRINT"
}

def clean_and_evaluate_expression(expr_str, memory):
    """Safely solves math using our local language memory bank."""
    # Convert Indian localized numbers (०-९) to standard English numbers (0-9)
    indian_digits = "०१२३४५६७८९"
    for i, digit in enumerate(indian_digits):
        expr_str = expr_str.replace(digit, str(i))

    # Replace variable names with their actual numerical values from memory
    for var_name, var_val in memory.items():
        expr_str = re.sub(r'\b' + re.escape(var_name) + r'\b', str(var_val), expr_str)

    try:
        return eval(expr_str)
    except Exception:
        return 0

def running_bharat_compiler_v2(script_text):
    memory = {}
    lines = script_text.strip().split('\n')

    for line in lines:
        line = line.strip()
        if not line or line.startswith("#"):
            continue

        words = line.split()
        translated_words = [KEYWORDS.get(w, w) for w in words]
        translated_line = " ".join(translated_words)

        # 1. Handling Variable Assignment + Math Operations (LET x = y + z)
        if translated_line.startswith("LET"):
            match = re.match(r"LET\s+(\w+)\s*=\s*(.*)", translated_line)
            if match:
                var_name, expression = match.groups()
                # Run the math equation through our calculator evaluator
                memory[var_name] = clean_and_evaluate_expression(expression, memory)

        # 2. Handling Conditionals (IF x < y PRINT "message")
        elif translated_line.startswith("IF"):
            match = re.match(r"IF\s+(.*?)\s+PRINT\s*\"(.*)\"", translated_line)
            if match:
                condition, message = match.groups()
                if clean_and_evaluate_expression(condition, memory):
                    print(message)

print("Engine v2.0 Upgraded with Expression Math Math!")
```

```python
import re

# Keyword Map including Loops
KEYWORDS = {
    "मान": "LET", "மாறி": "LET", "రఖో": "LET",
    "यदि": "IF", "எனில்": "IF", "అయితే": "IF",
    "जबतक": "WHILE", "வரை": "WHILE", "వరకు": "WHILE",
    "दिखाओ": "PRINT", "அச்சிடு": "PRINT", "చూపించు": "PRINT"
}

def clean_and_evaluate_expression(expr_str, memory):
    indian_digits = "०१२३४५६७८९"
    for i, digit in enumerate(indian_digits):
        expr_str = expr_str.replace(digit, str(i))
    for var_name, var_val in memory.items():
        expr_str = re.sub(r'\b' + re.escape(var_name) + r'\b', str(var_val), expr_str)
    try:
        return eval(expr_str)
    except Exception:
        return 0

def running_bharat_compiler_v3(script_text):
    memory = {}
    # Split code by lines
    lines = [line.strip() for line in script_text.strip().split('\n') if line.strip() and not line.strip().startswith("#")]

    pc = 0  # Program Counter to track which line we are executing
    loop_stack = []  # Tracks open loops

    while pc < len(lines):
        line = lines[pc]
        words = line.split()
        translated_words = [KEYWORDS.get(w, w) for w in words]
        translated_line = " ".join(translated_words)

        # 1. VARIABLE LET
        if translated_line.startswith("LET"):
            match = re.match(r"LET\s+(\w+)\s*=\s*(.*)", translated_line)
            if match:
                var_name, expression = match.groups()
                memory[var_name] = clean_and_evaluate_expression(expression, memory)
            pc += 1

        # 2. CONDITION IF
        elif translated_line.startswith("IF"):
            match = re.match(r"IF\s+(.*?)\s+PRINT\s*\"(.*)\"", translated_line)
            if match:
                condition, message = match.groups()
                if clean_and_evaluate_expression(condition, memory):
                    print(message)
            pc += 1

        # 3. LOOPS (WHILE condition DO)
        elif translated_line.startswith("WHILE"):
            match = re.match(r"WHILE\s+(.*?)\s+करो", translated_line) # 'करो' signals loop start
            if match:
                condition = match.groups()[0]
                if clean_and_evaluate_expression(condition, memory):
                    # If condition is true, push loop info to stack and continue inside loop
                    if not loop_stack or loop_stack[-1]['start_line'] != pc:
                        loop_stack.append({'start_line': pc, 'condition': condition})
                    pc += 1
                else:
                    # If condition is false, skip past the loop block down to 'बन्द'
                    scan_pc = pc + 1
                    open_loops = 1
                    while scan_pc < len(lines):
                        if "करो" in lines[scan_pc]: open_loops += 1
                        if "बन्द" in lines[scan_pc]: open_loops -= 1
                        if open_loops == 0:
                            break
                        scan_pc += 1
                    pc = scan_pc + 1
            else:
                pc += 1

        # 4. LOOP END (बन्द)
        elif translated_line.startswith("बन्द") or translated_line.startswith("முடி"):
            if loop_stack:
                # Jump back up to the header line of the loop to re-evaluate the condition
                pc = loop_stack[-1]['start_line']
                loop_stack.pop() # Clear stack frame to prevent infinite compilation loop accumulation
            else:
                pc += 1
        else:
            pc += 1

print("Engine v3.0 Ready: Math, Conditions, and Loops are Online!")
```

```python
loop_script_hindi = """
मान गिनती = १
जबतक गिनती <= ३ करो
    दिखाओ "लूप चल रहा है, वर्तमान संख्या:"
    दिखाओ गिनती
    मान गिनती = गिनती + १
बन्द
"""

print("--- Testing Loop Iteration ---")
running_bharat_compiler_v3(loop_script_hindi)
```
